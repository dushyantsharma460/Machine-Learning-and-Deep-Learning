```python
# MNIST Dataset as a benchmark for Neural Networks 
# The MNIST dataset is used for benchmarking and training machine learning and deep learning algorithms, primarily for image classification and recognition tasks, especially handwritten digit recognition.

# Now we will visualize this dataset
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
print(f"Training data shape is {x_train.shape}")
```

```python
print(f"Training data shape is {x_test.shape}")
```

```python
plt.figure(figsize=(10, 2))
for i in range(10):
    plt.subplot(1, 10, i + 1)
    plt.imshow(x_train[i], cmap="gray")
    plt.axis("off")
    plt.title(str(y_train[i]))
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 2))
for i in range(100, 110):
    plt.subplot(1, 10, i + 1 -100)
    plt.imshow(x_train[i], cmap="gray")
    plt.axis("off")
    plt.title(str(y_train[i]))
plt.tight_layout()
plt.show()
```
