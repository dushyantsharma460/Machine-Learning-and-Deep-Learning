```python
# But i also do train test split data on Iris -> Iris Train + Iris Test

data1 = pd.read_excel("Iris-train.xlsx")
```

```python
x = data1.iloc[:,:-1]  # All row and all column except last column
```

```python
x
```

```python
x.values
```

```python
y = data1.iloc[:,-1]
```

```python
y
```

```python
model = GaussianNB()
```

```python
model.fit(x,y)
```

```python
prediction = model.predict([[5.4, 3, 4.5, 1.5]])
```

```python
prediction
```

```python
dataTest = pd.read_excel("Iris-test.xlsx")
```

```python
dataTest
```

```python
x_test = dataTest.iloc[:,:-1]
```

```python
whole_predictions = model.predict(x_test.values)
```

```python
whole_predictions
```

```python
y_test = dataTest.iloc[:,-1]
```

```python
actual = y_test.values
```

```python
actual
```

```python
count = 0
for i in range(0, len(actual)):
    if actual[i] == whole_predictions[i]:
        count = count + 1
print(f"Accuracy is: {(count * 100) / len(actual):.2f}%")
```
