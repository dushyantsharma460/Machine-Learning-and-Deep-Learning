```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("housing.csv")
```

```python
df['income_category'] = pd.cut(df['median_income'], bins = [0, 1.5, 3.0, 4.5, 6.0, np.inf], labels = [1, 2, 3, 4, 5]) 
```

```python
df
```

```python
# Stratified sampling needs a categorical column to decide how to split.
# Stratified sampling needs groups (categories) to balance the split.
# If you try to stratify directly on a continuous number like median_income, almost every row is unique → no meaningful “strata” to preserve.
# median_income   →   income_category
# ----------------------------------
# 0.8             →   1   (low income)
# 2.5             →   2
# 3.7             →   3
# 5.2             →   4
# 7.8             →   5   (high income)
# ocean_proximity tells you about location (like “INLAND”, “NEAR OCEAN”, etc.).
# The dataset’s target (house value) is strongly related to income, not just proximity.
# If you only stratify on ocean_proximity, your train/test sets may not preserve the income distribution — which is more critical for predicting house prices.

# Spread median_income equally both in training and testing set 

from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_category']):     # df['income_category'] = the column to use for stratification
    strat_train_set = df.loc[train_index]                                  # df['income_category'] -> Matlab df['income_category'] sab ma hona chaiya
    strat_test_set = df.loc[test_index]                                    # Matlab income category wala column sab ma aana chaiya both train and test
```

```python
# Lets remove the income cat column 
for sett in (strat_train_set, strat_test_set):
    sett.drop("income_category", axis=1, inplace=True)
```

```python
# Before cleaning data we will first seprate the label from the data
housing = strat_train_set.drop("median_house_value", axis=1)
housing_label = strat_train_set["median_house_value"].copy()
```

```python
housing = housing.drop("ocean_proximity", axis=1)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
mypipeline = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
     ("standardize", StandardScaler()),
     
])
```

```python
# How It Works
# • The pipeline takes a list of steps as (name, transformer) pairs. like ("standardize", StandardScaler())
# • Names must be unique and should not contain double underscores __ .
# • All intermediate steps must be transformers (i.e., must implement fit_transform() ).
# • The final step can be either a transformer or a predictor.

mypipeline.fit_transform(housing)
```
