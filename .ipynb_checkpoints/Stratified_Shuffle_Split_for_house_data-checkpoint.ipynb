```python
import pandas as pd
```

```python
df = pd.read_csv("housing.csv")
```

```python
df
```

```python
import numpy as np
```

```python
# We have split data successfully but not properly. for ex :-  my training set not include values NEAR BAY(ocean_proximity)
# So, how it give correct predition for the test data 
# To ensure that important characteristics of the population are well represented in both the training and test sets, we use stratified sampling.
# I will use Strata (Strata is sub group of data)
# In my data median income is strong predictor house prices

# Adding new strata for median_income

df['income_category'] = pd.cut(df['median_income'], bins = [0, 1.5, 3.0, 4.5, 6.0, np.inf], labels = [1, 2, 3, 4, 5]) 
```

```python
import matplotlib.pyplot as plt
df["income_category"].value_counts().sort_index().plot.bar(rot=0, grid=True)  # rot -> rotation 
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_category']):     # df['income_category'] = the column to use for stratification
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
strat_test_set
```

```python
import matplotlib.pyplot as plt
strat_train_set["income_category"].value_counts().sort_index().plot.bar(rot=0, grid=True)  # rot -> rotation for labels
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
plt.show()
```

```python
import matplotlib.pyplot as plt
strat_test_set["income_category"].value_counts().sort_index().plot.bar(rot=0, grid=True)  # rot -> rotation for labels
plt.title("Income Categories Distribution")
plt.xlabel("Income Category")
plt.ylabel("Number of Instances")
plt.show()
```
